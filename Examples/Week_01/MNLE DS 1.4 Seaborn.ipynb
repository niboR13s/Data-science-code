# 1. SEABORN SCATTERPLOT & HISTPLOT

```
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# Seaborn offers powerful graphical features, yet simple to use
# https://seaborn.pydata.org/examples/index.html

# import libraries
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```
# Import Cancer data from the Scikit-learn library
# Scikit-learn contains a lot of sample data libraries:
# https://scikit-learn.org/stable/datasets.html
# Data set contains samples of cancer cells with many features and target feature 'malignant' (0) or 'benign' (1) (kwaadaardig/goedaardig)
# Typical classification problem
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
# Create a dataFrame named df_cancer with input/output data
# np.c_ class object translates slice objects to concatenation along the second axis
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```
# Check out the head of the dataframe
df_cancer.head()
```

```
# Check out the tail of the dataframe
df_cancer.tail()
```

```
# Plot scatter plot between mean area and mean smoothness
sns.scatterplot(x = 'mean area', y = 'mean smoothness', data = df_cancer)
```

```
# Plot including target class in color
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df_cancer)
```

```
# Let's print out countplot to know how many samples belong to class #0 and #1
sns.histplot(df_cancer['target'], bins = 2, label = "Count")
```

# 2. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS

```
# Plot the pairplot
# Pairplot without color indication of the target
#sns.pairplot(df_cancer, vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
# Pairplot with color indication of the target
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```

```
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (20, 10))
sns.heatmap(df_cancer.corr(), annot = True)
```

```
# plot the displot
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable.

sns.displot(df_cancer['mean radius'], bins = 25, kde=True, color = 'blue')
```

```
# separate the two target classes in two separate dataframes
class_0_df = df_cancer[ df_cancer['target']==0 ]
class_1_df = df_cancer[ df_cancer['target']==1 ]
```

```
# Plot the displot for both classes
plt.figure(figsize=(10, 7))
sns.displot(class_0_df['mean radius'], bins = 15, kde=True, color = 'blue')
sns.displot(class_1_df['mean radius'], bins = 15, kde=True, color = 'red')
plt.grid()
```

```
# Plot the displot of both
sns.displot(data=df_cancer, x='mean radius', bins = 25, hue = 'target', kde=True)
```
